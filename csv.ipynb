```python
import csv
f = open('bill.csv','a')
x = csv.writer(f)
h = ['Name','Price','Quantity','Total']
x.writerow(h)
data = [] 
n = int(input("enter n = "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("enter price = "))
    quantity = int(input("enter quantity = "))
    total = quantity*price
    info = [name,price,quantity,total]
    
    data.append(info)
    x.writerow(info)
print(data)
f.close()
```

```python
#dict writer
import csv
csv_coloumns = ['Name','Age','Address']
dict =   [
    {'Name':'Ram','Age':33,'Address':'Kathmandu'},
    {'Name':'Shyam','Age':33,'Address':'Kathmandu'}
    
]
try:
    with open('data.csv','w') as csvfiles:
        writer = csv.DictWriter(csvfiles,fieldnames = csv_coloumns)
        writer.writeheader()
        for data in dict:
            writer.writerow(data)
except IOError:
    print('I/O error')
```
