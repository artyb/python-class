```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus', 'html.parser')
response = requests.get(url)
response
```

```python
markup = response.text
xml_code = BeautifulSoup(markup, "lxml")
table_code = xml_code.table
table_code
```

```python
tags = table_code.find_all('tr')[1:]
tags
```

```python
tags[1].text.split('\n')
```

```python
output = []
for i in tags:
    x = (i.text).split('\n')[2:-1]
    output.append(x)
print(output)
```

```python
import csv
file = open('worldometer.csv','w')
x = csv.writer(file)
for i in output:
    x.writerow(i)
file.close()
```
